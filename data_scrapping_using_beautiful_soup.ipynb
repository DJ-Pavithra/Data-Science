```python
from urllib.request import urlopen
```

```python
url="https://www.geeksforgeeks.org/html/html-tables/"
```

```python
url_data=urlopen(url)
print(type(url_data))
```

```python
html=url_data.read()
print(html)
```

Parsing data

```python
from bs4 import BeautifulSoup as soup
```

```python
url_soup=soup(html,"html.parser")
print(url_soup)
```

```python
url_soup.h1
```

```python
url_soup.findAll('h1',{})
```

```python
inputs=url_soup.findAll('input',{'class':'gcse-search-input__wrapper'})
```
